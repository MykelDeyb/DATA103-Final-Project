```python
import pandas as pd
import numpy as np
# Load the dataset
df = pd.read_csv("C:/Users/acer/Downloads/PHL-PSA-NMS-2018-PUF/NMS PUF 2018  Individual.csv")

# Select relevant columns
df1 = df[['C305PROV', 'G701PROV']]
df1.columns = ['PreviousResidence_Code', 'CurrentResidence_Code']

# Filter rows
# Function to replace non-numeric values with NaN
def replace_non_numeric_with_na(series):
    return pd.to_numeric(series, errors='coerce')

# Apply the function to the specified columns
df1['PreviousResidence_Code'] = replace_non_numeric_with_na(df1['PreviousResidence_Code'])
df1['CurrentResidence_Code'] = replace_non_numeric_with_na(df1['CurrentResidence_Code'])
```

```python
main = df1[(df1['PreviousResidence_Code'] != df1['CurrentResidence_Code']) & 
            (df1['CurrentResidence_Code'].notna()) & 
            (df1['PreviousResidence_Code'].notna())].copy()
main['PreviousResidence_Code'] = main['PreviousResidence_Code'].astype(int)
main['CurrentResidence_Code'] = main['CurrentResidence_Code'].astype(int)
```
